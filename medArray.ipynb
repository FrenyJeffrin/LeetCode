# **CONTAINER WITH MOST WATER**

You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are (i, 0) and (i, height[i]).

Find two lines that together with the x-axis form a container, such that the container contains the most water.

Return the maximum amount of water a container can store.

Notice that you may not slant the container.

Example 1:


Input: height = [1,8,6,2,5,4,8,3,7]
Output: 49
Explanation: The above vertical lines are represented by array [1,8,6,2,5,4,8,3,7]. In this case, the max area of water (blue section) the container can contain is 49.
Example 2:

Input: height = [1,1]
Output: 1


Constraints:

n == height.length
2 <= n <= 105
0 <= height[i] <= 104

```python
from typing import List

class Solution:
    def maxArea(self, height: List[int]) -> int:
        i=0
        j=len(height) -1
        res=0

        while i<j:
            res= max(res,(j-i)*min(height[j],height[i]))
            if height[i]<height[j]:
                i +=1
            else:
                j -=1
        return res
```

```python
solution = Solution()
height_example1 = [1,8,6,2,5,4,8,3,7]
result_example1 = solution.maxArea(height_example1)
print(f"Input: height = {height_example1}")
print(f"Output: {result_example1}")
```

## **3** **SUM**

Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] such that i != j, i != k, and j != k, and nums[i] + nums[j] + nums[k] == 0.

Notice that the solution set must not contain duplicate triplets.



Example 1:

Input: nums = [-1,0,1,2,-1,-4]
Output: [[-1,-1,2],[-1,0,1]]
Explanation:
nums[0] + nums[1] + nums[2] = (-1) + 0 + 1 = 0.
nums[1] + nums[2] + nums[4] = 0 + 1 + (-1) = 0.
nums[0] + nums[3] + nums[4] = (-1) + 2 + (-1) = 0.
The distinct triplets are [-1,0,1] and [-1,-1,2].
Notice that the order of the output and the order of the triplets does not matter.
Example 2:

Input: nums = [0,1,1]
Output: []
Explanation: The only possible triplet does not sum up to 0.
Example 3:

Input: nums = [0,0,0]
Output: [[0,0,0]]
Explanation: The only possible triplet sums up to 0.


Constraints:

3 <= nums.length <= 3000
-105 <= nums[i] <= 105

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        nums.sort()   # Sort array
        n = len(nums)
        res = []

        for i in range(n - 2):
            # Skip duplicate
            if i > 0 and nums[i] == nums[i - 1]:
                continue

            left, right = i + 1, n - 1
            target = -nums[i]

            while left < right:
                s = nums[left] + nums[right]

                if s == target:
                    # Found a valid triplet
                    res.append([nums[i], nums[left], nums[right]])
                    left += 1
                    right -= 1

                    # Skip duplicates for left
                    while left < right and nums[left] == nums[left - 1]:
                        left += 1
                    # Skip duplicates for right
                    while left < right and nums[right] == nums[right + 1]:
                        right -= 1

                elif s < target:
                    left += 1
                else:
                    right -= 1

        return res
```

```python
solution = Solution()
nums_example1 = [-1,0,1,2,-1,-4]
result_example1_threesum = solution.threeSum(nums_example1)
print(f"Input: nums = {nums_example1}")
print(f"Output: {result_example1_threesum}")
```

Problem Statement :
We are given an array of integers nums.
Our job is to find all unique groups of 3 numbers (called triplets) from this array such that:
If you add the 3 numbers together, the result is 0.
Each number in the triplet must come from a different index in the array.
You are not allowed to return duplicates — meaning if the same 3 numbers appear again in another order, you only count them once.
The order of the triplets in your answer doesn’t matter.
Example 1:
Input: nums = [-1, 0, 1, 2, -1, -4]
Triplets that sum to zero:

[-1, 0, 1] → because -1 + 0 + 1 = 0
[-1, -1, 2] → because -1 + -1 + 2 = 0
Final Answer: [[-1, -1, 2], [-1, 0, 1]]

Intuition
USing Two pointers we can give the Optimal solution for this problem.
If you fix one number a = nums[i], the problem becomes: find two numbers b and c (from the remaining elements) such that b + c = -a. That is a two-sum problem.
If the array is sorted, two-sum can be solved with two pointers:
Move left forward to increase sum
Move right backward to decrease sum
Sorting also helps us easily skip duplicates.
Approach (Two pointers)
Idea
Sort nums. For each index i as the first element:

Use two pointers (left = i+1, right = n-1) to find pairs.
If nums[left] + nums[right] == -nums[i] → record triplet.
Skip duplicates for i, left, and right.
Two-pointer approach :
1) Sort the array
Sort nums in non-decreasing order.
Why: sorting gives order (smaller on left, larger on right) so pointer moves are predictable.
Benefit: makes it easy to increase/decrease pair sums and to detect/skip duplicates.
2) Fix one element nums[i] as the first of the triplet
For each i (0 .. n-3) treat a = nums[i] as the first element.
The problem reduces to: find b and c such that b + c = -a (a two-sum).
Only consider i up to n-3 because you need two more elements after i.
3) Use two pointers on the subarray right of i
Initialize: left = i + 1, right = n - 1.
Let target = -nums[i]. We want nums[left] + nums[right] == target.
Because the array is sorted:
If nums[left] + nums[right] < target, then left++ (increases pair-sum).
If nums[left] + nums[right] > target, then right-- (decreases pair-sum).
Continue while left < right. Every valid pair for this i will be found in linear time.
4) Skip duplicates carefully (avoid repeated triplets)
Before two-pointer scan:
If i > 0 and nums[i] == nums[i-1], continue (skip same a).
After finding a valid triplet:
Move both pointers: left += 1; right -= 1.
Then skip repeated values:
     while left < right and nums[left] == nums[left - 1]: left += 1`
     while left < right and nums[right] == nums[right + 1]: right -= 1
This ensures each unique triplet is added exactly once.
5) Early stopping optimization
If nums[i] > 0 then break the loop:
Because array is sorted, nums[i] and everything after it is ≥ nums[i], so sum of three positives cannot be 0.
6) Quick tips & edge cases
If n < 3: return [].
All zeros: algorithm finds [0,0,0] once and skips duplicates.
Large duplicates: duplicate skipping is essential to avoid repeated results and extra work.
Careful with pointer bounds when skipping duplicates after moving left/right.
Algorithm (Steps)
Edge case check

If len(nums) < 3: return [] (no triplets possible).
Sort the array

Sort nums in non-decreasing order.
Initialize result

res = [] to store all unique triplets.
Iterate through array

For i from 0 to n - 3:
Skip duplicates for the first element
If i > 0 and nums[i] == nums[i-1]: continue
Early stopping optimization
If nums[i] > 0: break (no further triplets possible)
Set target for two-sum
target = -nums[i]
Initialize two pointers
left = i + 1, right = n - 1
Two-pointer scan
While left < right:
Compute s = nums[left] + nums[right]
If s == target:
Append [nums[i], nums[left], nums[right]] to res
Move both pointers: left += 1, right -= 1
Skip duplicates:
while left < right and nums[left] == nums[left - 1]: left += 1
while left < right and nums[right] == nums[right + 1]: right -= 1
Else if s < target: left += 1 (need a larger sum)
Else (s > target): right -= 1 (need a smaller sum)
Return result

After the loop finishes, return res
